```python
def convert(x):
    if x == 'No Disorder':
        return 1
    elif x == 'Sleep Apnea':
        return 2
    elif x == 'Insomnia':
        return 3
```

```python
y=y.apply(convert)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled_df=pd.DataFrame(X_train_scaled, columns= X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

### 1st Approch

#### SVM

```python
from sklearn.svm import SVC
s_model=SVC()
s_model.fit(X_train_scaled,y_train)
```

```python
s_y_pred=s_model.predict(X_test_scaled)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,s_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,s_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,s_y_pred),annot= True,fmt='d')
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
k_model=KNeighborsClassifier(n_neighbors=1)
k_model.fit(X_train_scaled,y_train)
```

```python
k_y_pred=k_model.predict(X_test_scaled)
```

```python
error_rates=[]
k_values=range(1,40)
for k in k_values:
    k_model=KNeighborsClassifier(n_neighbors=k)
    k_model.fit(X_train_scaled,y_train)
    k_y_pred=k_model.predict(X_test_scaled)
    error_rate=1- metrics.accuracy_score(y_test,k_y_pred)
    error_rates.append(error_rate)

optimal_k=k_values[np.argmin(error_rates)]
print(f"optimal value of k is {optimal_k}")
print(f"error rate at optimal k is {min(error_rates)}")
```

```python
k_model=KNeighborsClassifier(n_neighbors=24)
k_model.fit(X_train_scaled,y_train)
```

```python
k_y_pred=k_model.predict(X_test_scaled)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,k_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,k_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,k_y_pred),annot= True,fmt='d')
```

#### DT

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train_scaled,y_train)
```

```python
dt_y_pred=dt_model.predict(X_test_scaled)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,dt_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,dt_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,dt_y_pred),annot= True,fmt='d')
```

#### RF

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(X_train_scaled,y_train)
```

```python
rf_y_pred=rf_model.predict(X_test_scaled)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,rf_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,rf_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,rf_y_pred),annot= True,fmt='d')
```

```python
o=pd.DataFrame({
    'Model' : ['SVM', 'KNN', 'DT', 'RF'],
    'Score' : [cross_val_score(SVC(),X,y,cv=5).mean(),
               cross_val_score(KNeighborsClassifier(),X,y,cv=5).mean(),
              cross_val_score(DecisionTreeClassifier(),X,y,cv=5).mean(),
               cross_val_score(RandomForestClassifier(),X,y,cv=5).mean()]

})
o.sort_values(by='Score', ascending= False)

```

```python
o=pd.DataFrame({
    'Model' : ['SVM', 'KNN', 'DT', 'RF'],
    'Score' : [s_model.score(X_train_scaled,y_train), k_model.score(X_train_scaled,y_train),
              dt_model.score(X_train_scaled,y_train), rf_model.score(X_train_scaled,y_train)]

})
o.sort_values(by='Score', ascending= False)
```

### 2nd Approach

#### Feature selection-RFE (using SVM model)

```python
from sklearn.feature_selection import RFE

svm_model = SVC(kernel='linear', random_state=42)

# Perform RFE to select top 4 features
rfe = RFE(estimator=svm_model, n_features_to_select=6)
rfe.fit(X_train_scaled, y_train)

# Get selected features from RFE
selected_features = X_train_scaled_df.columns[rfe.support_]
print("Top 6 Features Selected by RFE:", selected_features)
```

#### Feature importance ( using RandomForestâ€™s .feature_importances_) --> to cross-check if the right features are being selected.

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train)

# Get feature importance
feature_importance = rf_model.feature_importances_

# Display feature importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
rfe_features = ['Occupation', 'Sleep Duration', 'Stress Level', 'Blood Pressure', 'Heart Rate', 'Daily Steps']
rf_features = ['Blood Pressure', 'Sleep Duration', 'Age', 'Occupation', 'Physical Activity Level', 'Heart Rate']

selected_features = list(set(rfe_features) & set(rf_features))
print("Features Selected:", selected_features)

# Create datasets with selected features
X_train_selected = X_train_scaled_df[selected_features]
X_test_selected = X_test_scaled_df[selected_features]
```

#### SVM

```python
from sklearn.model_selection import GridSearchCV
params={'C': [0.1, 1, 10], 'gamma': [0.001, 0.01, 0.1], 'kernel' : ['linear', 'rbf']}
s_gsv=GridSearchCV(SVC(),params,cv=5,scoring='accuracy')
s_gsv.fit(X_train_selected,y_train)
```

```python
s_model=SVC(**s_gsv.best_params_)
```

```python
s_model.fit(X_train_selected,y_train)
```

```python
s_y_pred=s_model.predict(X_test_selected)
```

```python

print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,s_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,s_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,s_y_pred),annot= True,fmt='d')
```

#### KNN

```python

params = {
    'n_neighbors': range(1, 40),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

k_gsv= GridSearchCV(KNeighborsClassifier(), params, scoring='accuracy', cv=5)
k_gsv.fit(X_train_selected, y_train)
```

```python
k_model=KNeighborsClassifier(**k_gsv.best_params_)
k_model.fit(X_train_selected, y_train)
```

```python
k_y_pred = k_model.predict(X_test_selected)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,k_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,k_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,k_y_pred),annot= True,fmt='d')
```

#### DT

```python

params = {'max_depth': [2, 4, 6, 8], 'min_samples_split': [2, 4, 6, 8]}
dt_gsv=GridSearchCV(DecisionTreeClassifier(),params,cv=5,scoring='accuracy')
dt_gsv.fit(X_train_selected,y_train)
```

```python

dt_model=DecisionTreeClassifier(**dt_gsv.best_params_)
dt_model.fit(X_train_selected,y_train)
```

```python
dt_y_pred=dt_model.predict(X_test_selected)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,dt_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,dt_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,dt_y_pred),annot= True,fmt='d')
```

#### RF

```python

params = {'max_depth': [4, 8, 12], 'n_estimators': [100, 200, 300]}
rf_gsv=GridSearchCV(RandomForestClassifier(),params,cv=5,scoring='accuracy')
rf_gsv.fit(X_train_selected,y_train)
```

```python

rf_model=RandomForestClassifier(**rf_gsv.best_params_)
rf_model.fit(X_train_selected,y_train)
```

```python
rf_y_pred=rf_model.predict(X_test_selected)
```

```python
print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,rf_y_pred)}")
print("Classification report")
print(metrics.classification_report(y_test,rf_y_pred))
sns.heatmap(metrics.confusion_matrix(y_test,rf_y_pred),annot= True,fmt='d')
```
